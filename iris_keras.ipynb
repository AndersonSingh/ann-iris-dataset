```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.utils import shuffle

from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from keras.callbacks import EarlyStopping

# read data as a panda dataframe and set the headers appropriately.
iris_df = pd.read_csv('iris_data.csv', header = 0, names = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class'])

# shuffle the dataframe since the rows are ordered. 
iris_df = shuffle(iris_df)


# explore the data set.
# print(iris_df.head(5))
# print(iris_df.describe())
```

```python
# get the predictors into a numpy array for keras. 
# dont want to deal to pandas, just numpy.
np_predictors = iris_df.as_matrix(columns=['sepal length', 'sepal width', 'petal length', 'petal width'])

# get the classes, and put them in a 1D array.
np_classes = iris_df.as_matrix(columns=['class'])
np_classes = np_classes.ravel()

# convert the labels to numeric representation.
labelEncoder = LabelEncoder()

# use keras util function to convert labels to numeric matrix.
targets = to_categorical(labelEncoder.fit_transform(np_classes))

# we have our predictors and our targets for them. 

# use a sequential model. 
model = Sequential()

# get the number of input cols. 
n_cols = np_predictors.shape[1]
input_shape = (n_cols, )

# add 1 hidden layer
model.add(Dense(8, activation='relu', input_shape=input_shape))

# add 3 nodes to ouput layer.
model.add(Dense(3, activation='softmax'))

# implement earlystopping to stop training model when quality does not improve.
earlyStopping= EarlyStopping(patience=2)

# compile and fit model. 
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(np_predictors, targets, epochs=200, validation_split=0.3, callbacks=[earlyStopping])
```
